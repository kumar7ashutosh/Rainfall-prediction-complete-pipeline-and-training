```python
df.rename(columns={'humidity ':'humidity', 'pressure ':'pressure', 'cloud ':'cloud', '         winddirection':'winddirection'},inplace=True)
```

```python
num_features=df.select_dtypes(exclude='object').columns
cat_features=df.select_dtypes(include='object').columns
```

```python
num_features
```

```python
ss=StandardScaler()
le=LabelEncoder()
df['rainfall']=le.fit_transform(df['rainfall'])
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_pipeline=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('Scaling',ss)
])
cat_pipeline=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
])
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ('num',num_pipeline,num_features),
        ('cat',cat_pipeline,df['rainfall'])
    ],remainder='passthrough'
)
```

```python
pipeline=Pipeline(steps=[('Preprocessor',preprocessor)])
```

```python
from imblearn.combine import SMOTETomek
```

```python
x=df.drop(columns=['rainfall'],axis=1)
y=df['rainfall']

num_features=x.select_dtypes(exclude='object').columns
cat_features=x.select_dtypes(include='object').columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train=pipeline.fit_transform(x_train)
```

```python
x_test=pipeline.transform(x_test)
```

```python
smt=SMOTETomek(random_state=42)
```

```python
x_train_res,y_train_res=smt.fit_resample(x_train,y_train)
```

```python
le.inverse_transform([0,1])
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
def evaluate_model(y_hat,y_pred):
  return accuracy_score(y_hat,y_pred),f1_score(y_hat,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
models={
    'lc':LogisticRegression(),
    'knn':KNeighborsClassifier(),
    'svc':SVC(),
    'dtc':DecisionTreeClassifier(),
    'rfc':RandomForestClassifier(),
    'abc':AdaBoostClassifier(),
    'cbc':CatBoostClassifier(),
    'xbc':XGBClassifier()
}
```

```python
params = {
    'lc': {
        "C": [0.01, 0.1, 1, 10],
        "solver": ["lbfgs", "liblinear"]
    },
    "knn": {
        "n_neighbors": [3, 5, 7, 9],
        "weights": ["uniform", "distance"]
    },
    "dtc": {
        "max_depth": [3, 5, 10, None],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4]
    },
    "rfc": {
        "n_estimators": [100, 200],
        "max_depth": [5, 10, 15, None],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4]
    },
    "xbc": {
        "n_estimators": [100, 200],
        "learning_rate": [0.01, 0.05, 0.1],
        "max_depth": [3, 5, 7]
    },
    "cbc": {
        "iterations": [100, 200],
        "depth": [4, 6],
        "learning_rate": [0.01,0.1],
        "l2_leaf_reg": [1,5]
    },
    "abc": {
        "n_estimators": [50, 100],
        "learning_rate": [ 0.1, 1]
    },
    "svc": {
        "C": [0.1, 1, 10],
        "kernel": ["linear", "rbf"],

    }
}
```

```python
report={}
best_model=None
best_score=-np.inf
for name,model in models.items():
  print(f'Training {name}-----')
  param_grid=params[name]
  if len(param_grid)>0:
    search=RandomizedSearchCV(model,param_grid,n_iter=5,scoring='f1_weighted',cv=3,verbose=1,random_state=42,n_jobs=-1)
    search.fit(x_train_res,y_train_res)
    best_estimator=search.best_estimator_
    print(f'best params for {name}:{search.best_params_}')
  else:
    best_estimator=model.fit(x_train_res,y_train_res)
  y_pred=best_estimator.predict(x_test)

  acc,f1=evaluate_model(y_test,y_pred)
  report[name]={'accuracy':acc,'f1':f1}
  if f1>best_score:
    best_score=f1
    best_model=best_estimator
```

```python
print(best_model,best_score)
```
